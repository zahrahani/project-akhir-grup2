```python
class Kursus:
  def __init__(self, nama_kursus, kode_kursus, durasi, kapasitas):
    self.nama_kursus = nama_kursus
    self.kode_kursus = kode_kursus
    self.durasi = durasi
    self.kapasitas = kapasitas
    self.peserta = 0

  def tambah_peserta(self):
    if self.peserta < self.kapasitas:
       self.peserta += 1
       print()
       print("\033[1mPeserta berhasil ditambahkan!\033[0m")
       print("-" * 36)
       print()
    else:
        print("Kapasitas kursus sudah penuh,")
        print("Peserta gagal ditambahkan!")
        print("-" * 36)
        print()

  def info_kursus(self):
    print(f"Nama Kursus     : {self.nama_kursus}")
    print(f"Kode Kursus     : {self.kode_kursus}")
    print(f"Durasi          : {self.durasi} minggu")
    print(f"Kapasitas       : {self.kapasitas}")
    print(f"Jumlah Peserta  : {self.peserta}")

daftar_kursus = []
data_kursus = {}

def cari_kursus(kode):
  if kode in data_kursus:
     data_kursus[kode].info_kursus()
  else:
     print(f"Kode kursus {kode} tidak ditemukan!")
     print("Pastikan penulisan kode kursus benar!")

def tampilkan_semua_kursus():
  if not daftar_kursus:
     print("Tidak ada kursus yang tersedia!")
     print("-" * 36)
     print()
  else:
    for kursus in daftar_kursus:
        kursus.info_kursus()
        print("          <<<<<<<<>>>>>>>>          ")
        print()

while True:
  print("\033[1m---------- Group 2 Course ----------\033[0m")
  print("-" * 36)
  print("Menu:")
  print("1. Tambah Kursus")
  print("2. Tambah Peserta")
  print("3. Cari Kursus")
  print("4. Tampilkan Semua Kursus")
  print("5. Keluar")
  print("-" * 36)

  while True:
    pilihan = input("Pilih Menu (1/2/3/4/5): ")
    if not pilihan:
       print("ERROR : Menu harus dipilih!")
       print()
       continue

    try:
      pilihan = int(pilihan)
    except ValueError:
      print("ERROR : Menu harus dipilih dengan angka!")
      print()
      continue
    break

  # penanganan menu 1
  if pilihan == 1:
     print("\033[1m        Menambah Kursus Baru        \033[0m")
     print()

     while True:
      nama = input("Nama Kursus       : ")
      if not nama:
         print("Nama kursus harus diisi!")
         continue
      break

     while True:
      kode = input("Kode Kursus       : ")
      if not kode:
         print("Kode kursus harus diisi!")
         continue
      break

     while True:
      durasi = input("Durasi (minggu)   : ")
      if not durasi:
         print("Durasi kursus harus diisi!")
         continue
      try:
        durasi = int(durasi)
        if durasi <= 0:
           print("Durasi harus > 0!")
           continue
        break
      except:
        print("Durasi harus berupa angka!")

     while True:
      kapasitas = input("Kapasitas Kursus  : ")
      if not kapasitas:
         print("Kapasitas harus diisi!")
         continue
      try:
        kapasitas = int(kapasitas)
        if kapasitas <= 0:
           print("Kapasitas harus > 0!")
           continue
        break
      except:
        print("Kapasitas harus berupa angka!")

     if kode in data_kursus:
        print(f"\033[1mTidak bisa menambahkan kursus {nama} lagi\033[0m")
        print(f"Kursus {nama} sudah tersedia!")
        print()
     else:
        kursus_baru = Kursus(nama, kode, durasi, kapasitas)
        daftar_kursus.append(kursus_baru)
        data_kursus[kode] = kursus_baru
        print()
        print(f"\033[1mKursus {nama} berhasil ditambahkan!\033[0m")
        print("-" * 36)
        print()

  # penanganan menu 2
  elif pilihan == 2:
       print("\033[1m    Menambah Peserta Baru Kursus    \033[0m")
       print()

       while True:
        kode = input("Kode Kursus : ")
        if not kode:
           print("Kode kursus harus diisi!")
           continue
        elif kode in data_kursus:
             data_kursus[kode].tambah_peserta()
        else:
           print(f"Kursus {kode} tidak ditemukan!")
           print("Harap masukkan kode dengan benar!")
           print()
           continue
        break
        print("-" * 36)
        print()

  # penanganan menu 3
  elif pilihan == 3:
       print("\033[1mMencari Kursus yang Tersedia\033[0m")
       print()

       kode = input("Kode Kursus : ")
       if not kode:
          print("Program tidak bisa mencari kursus,")
          print("Harap masukkan kode kursus!")
       else:
        cari_kursus(kode)
        print("-" * 36)
        print()

  # penanganan menu 4
  elif pilihan == 4:
       print("\033[1mSemua Kursus di Group 2 Course\033[0m")
       print()

       tampilkan_semua_kursus()
       print("-" * 36)
       print()

  # penanganan menu 5
  elif pilihan == 5:
       break

  # penanganan invalid input
  else:
    print("\033[1mInput tidak valid!\033[0m")
    print("-" * 36)
    print()
```
