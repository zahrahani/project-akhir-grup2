```python
# Program Manajemen Kursus
# dengan : PBO, Struktur Data, Fungsi, dan Try-Except
# dibuat oleh : Group 2

# kelas kursus dgn atribut nama kursus, kode kursus, durasi, kapasitas
class Kursus:
    def __init__(self, nama_kursus, kode_kursus, durasi, kapasitas):
        self.nama_kursus = nama_kursus
        self.kode_kursus = kode_kursus
        self.durasi = durasi
        self.kapasitas = kapasitas
        self.peserta = 0

# metode tambah peserta
    def tambah_peserta(self):
        if self.peserta < self.kapasitas:
            self.peserta += 1
            print()
            print("Peserta berhasil ditambahkan!")
            print()
        else:
            print()
            print("Peserta gagal ditambahkan!")
            print("Kapasitas kursus sudah penuh!")
            print()

# metode info kursus
    def info_kursus(self):
        print(f"Nama Kursus    : {self.nama_kursus}")
        print(f"Kode Kursus    : {self.kode_kursus}")
        print(f"Durasi         : {self.durasi} minggu")
        print(f"Kapasitas      : {self.kapasitas}")
        print(f"Jumlah Peserta : {self.peserta}")
        print()

# list data
daftar_kursus = []
data_kursus = {}

# menampilkan semua kursus
def tampilkan_semua_kursus():
    if not daftar_kursus:
        print("Tidak ada kursus yang tersedia!")
    else:
        for kursus in daftar_kursus:
            kursus.info_kursus()
            print("         <<<<<<<<<o>>>>>>>>>         ")
            print()

# mencari kursus pake kode
def cari_kursus(kode):
    if kode in data_kursus:
        data_kursus[kode].info_kursus()
    else:
        print(f"Kursus {kode} tidak ditemukan!")
        print("Pastikan penulisan kode kursus benar!")
        print()

print("\033[1m----------- Grub 2 Course -----------\033[0m")
print("-" * 37)
print("Menu:")
print("1. Tambah Kursus")
print("2. Tambah Peserta")
print("3. Cari Kursus")
print("4. Tampilkan Semua Kursus")
print("5. Keluar")
print("-" * 37)

while True:
    print()
    pilihan = input("Pilih Menu (1/2/3/4/5): ")

# menambahkan kursus
    if pilihan == '1':
        print("\033[1mMenambah Kursus Baru\033[0m")
        print()
        nama = input("Nama Kursus       : ")
        kode = input("Kode Kursus       : ")
        try:
            durasi = int(input("Durasi (minggu)   : "))
            kapasitas = int(input("Kapasitas Kursus  : "))

            kursus_baru = Kursus(nama, kode, durasi, kapasitas)
            daftar_kursus.append(kursus_baru)
            data_kursus[kode] = kursus_baru

            print()
            print(f"Kursus {nama} berhasil ditambahkan!")
            print()
            print("-" * 37)

        except ValueError:
            print()
            print("Durasi / Kapasitas harus berupa angka!")
            print()
            print("-" * 37)

# menambahkan peserta
    elif pilihan == '2':
        print("\033[1mMenambah Peserta Baru Kursus\033[0m")
        print()
        kode = input("Kode Kursus: ")
        if kode in data_kursus:
            data_kursus[kode].tambah_peserta()
            print("-" * 37)
        else:
            print()
            print(f"Kursus {kode} tidak ditemukan!")
            print()
            print("-" * 37)

# mencari kursus pake kode
    elif pilihan == '3':
        print("\033[1mMencari Kursus yang Tersedia\033[0m")
        print()
        kode = input("Kode kursus yang ingin dicari: ")
        print()
        cari_kursus(kode)
        print("-" * 37)

# menampilkan semua kursus
    elif pilihan == '4':
        print("\033[1mSemua Kursus di G2 Course\033[0m")
        print()
        tampilkan_semua_kursus()
        print("-" * 37)

    elif pilihan == '5':
        break

    else:
        print("Pilihan tidak valid!")
```
